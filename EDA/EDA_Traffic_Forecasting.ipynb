```python
with h5py.File('/content/gdrive/My Drive/DATASET_GWAR/metr-la.h5','r') as hdf:
  ls = list(hdf.keys())
  print('List of Datasets in this file are: ',ls)
  data = hdf.get('df')
  print('Type of Data in df is: ', type(data))
  print('Keys in df Group are : ', list(data.keys()))
  print('Data in Axis 0 : \n', np.array(data['axis0']))
  sensor_id_vector_h5 = np.array(data['axis0'])
  # Converting Sensor IDs to Pandas DF
  sensor_id_df = pd.DataFrame(sensor_id_vector_h5)
  print('Shape of Axis 0 : ', np.array(data['axis0']).shape)
  print('Data in Axis 1 : \n', np.array(data['axis1']))
  print('Shape of Axis 1 : ', np.array(data['axis1']).shape)
  time_sequences_vector_h5 = np.array(data['axis1'])
  # Converting Time Sequences to Panas DF
  time_sequences_df = pd.DataFrame(time_sequences_vector_h5)
  print('Data in Block 0 Items : \n', np.array(data['block0_items']))
  print('Shape of Block 0 Items : ', np.array(data['block0_items']).shape)
  print('Data in Block 0 Values : \n', np.array(data['block0_values']))
  adj_mat_h5 = np.array(data['block0_values'])
  # Converting Sensor IDs to Pandas DF
  adj_mat_df = pd.DataFrame(adj_mat_h5)
  print('Shape of Block 0 Values : ', np.array(data['block0_values']).shape)
```
